```python
import numpy as np
import torch
from sklearn.model_selection import train_test_split

# Assuming X is your input array and Y is your output vector

# Generate sample data (replace this with your actual data)
X = np.random.rand(30, 5)  # Example input array of shape (100, 5)
Y = np.random.rand(30,1)     # Example output vector of shape (100,)

X = torch.tensor(X)
Y = torch.tensor(Y)
print(X)
```

```python

def top_split_y(X, Y, percent):
    # Calculate the number of samples for test set
    a = 0
    if isinstance(X, torch.Tensor):
        X = X.numpy()
        a = 1
    if isinstance(Y, torch.Tensor):
        Y = Y.numpy()
    print('flag 1')
    print(X.shape)
    print(Y.shape)
    test_size = int(percent/100 * len(Y))

    # Sort Y and get indices of the top 15% values
    sorted_indices = np.argsort(Y[:,0])[::-1][:test_size]
    print('the sorted indices is')
    print(sorted_indices)
    # Create test set
    X_test = X[sorted_indices,:]
    Y_test = Y[sorted_indices,:]
    print('flag 2')
    print(X.shape)
    print(Y.shape)
    # Create train set
    X_train = np.delete(X, sorted_indices, axis=0)
    Y_train = np.delete(Y, sorted_indices, axis=0)
    print('flag 3')
    print(X.shape)
    print(Y.shape)
    if a ==1:
        X_test = torch.tensor(X_test)
        Y_test = torch.tensor(Y_test)
        X_train = torch.tensor(X_train)
        Y_train = torch.tensor(Y_train)
    print('flag 4')
    print(X.shape)
    print(Y.shape)
    return X_test, Y_test, X_train, Y_train
# Now, X_train, X_test, Y_train, Y_test contain the split data

X_test, Y_test, X_train, Y_train = top_split_y(X,Y,15)

print('flag 5')
print(Y_test)
print(Y_train)
```

```python
import numpy as np
import torch

# Create a NumPy array
numpy_array = np.array([1, 2, 3])

# Convert NumPy array to PyTorch tensor
torch_tensor = torch.tensor(numpy_array)

# Perform some operations on the original tensor
result = torch_tensor * 2

# Create a new tensor by copying data from the original tensor
new_tensor = torch.tensor(torch_tensor)

# Check if they share the same computational graph
print(torch_tensor.is_leaf)  # Output: True
print(new_tensor.is_leaf)     # Output: True
print(result.is_leaf)         # Output: False
```
