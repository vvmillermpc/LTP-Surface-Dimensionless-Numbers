```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Create the plot
fig, ax1 = plt.subplots()

# Plot the data
ax1.plot(x, y, marker='o')
ax1.set_xlabel('Primary X-axis labels')

# Create a second x-axis below the first
ax2 = ax1.twiny()
ax2.xaxis.set_ticks_position('bottom')  # Set the position of the second x-axis
ax2.xaxis.set_label_position('bottom')  # Set the position of the second x-axis label
ax2.spines['bottom'].set_position(('outward', 40))  # Offset the second x-axis

# Set the labels for the second x-axis
ax2.set_xlabel('Secondary X-axis labels')
ax2.set_xticks(x)
ax2.set_xticklabels(['A', 'B', 'C', 'D', 'E'])

# Show the plot
plt.show()
```
