```python

file =  r"C:\Users\vvmil\Documents\Python_Vmil\Jupyter_Notebooks\Plasma_He_calcs\data\data_packing\data_2cm_23mm_packing\Mesbah-DBD\OES_20230508154644\spec_average_20230508155000_subBKG_0.30f_100ns_1khz_12kv_36.9.txt"

import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file using pandas
df = pd.read_csv(file, sep = '\s+', header=None)


y = df[1].to_numpy()
x = df[0].to_numpy()
index_706 = np.abs(x - 706).argmin()
index_728 = np.abs(x - 728).argmin()

x_706, y_706 = x[index_706], y[index_706]
x_728, y_728 = x[index_728], y[index_728]



# Plot the data


font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 12}
matplotlib.rc('font', **font)
plt.plot(df[0][850:], df[1][850:])
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
#plt.title('Plot Title')


plt.annotate('706 nm', xy=(x_706, y_706), xytext=(x_706 + 2, y_706 + 1000),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.annotate('728 nm', xy=(x_728, y_728), xytext=(x_728 -15  , y_728 + 1000),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.savefig('oes_spectra.jpeg')


plt.show()

```
